#***Creating And Inserting/Removing From An Array:-***

```python
a = [1,3]
print(a)
print(a[-1])
a.append(4)
a.append(10)
print(a)
a.insert(1, 2)
print(a)
a.remove(10)
print(a)
```

#***Combining 2 Arrays:-***

```python
b = [1,2,3,4,]
c = [5,6,7,8]
d = b + c
print(d)
```

#***User Input For Arrays:-***

```python
print('Enter names of your family members')
Names = input().split()
print(Names)
print('Enter their age')
Age = list(map(int, input().split()))
print(Age)
A = dict(zip(Names,Age))
print(A)
```
