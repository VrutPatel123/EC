#***1. Obtaining Histogram:-***

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
image_path = list(uploaded.keys())[0]
print("Uploaded file:", image_path)

def compute_grayscale_histogram(image_path):
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    hist_gray = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

    plt.figure(figsize=(3, 3))
    plt.title("Histogram")
    plt.xlabel("Intensity")
    plt.ylabel("Frequency")
    plt.plot(hist_gray, color='black')
    plt.xlim([0, 256])
    plt.show()

compute_grayscale_histogram(image_path)
```

#***2. Equilizing Histogram:-***

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
image_path = list(uploaded.keys())[0]

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
equalized_image = cv2.equalizeHist(image)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Equalized Image")
plt.imshow(equalized_image, cmap='gray')
plt.axis("off")

plt.show()
```

#***3. Edge Detection:-***

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
image_path = list(uploaded.keys())[0]

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(image, 100, 200)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Edge Detection")
plt.imshow(edges, cmap='gray')
plt.axis("off")

plt.show()
```

#***4. Image Segmentation:-***

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
image = cv2.imread(list(uploaded.keys())[0], cv2.IMREAD_GRAYSCALE)
_, segmented = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(9, 6))  # Increase figure size
plt.subplot(1, 2, 1), plt.imshow(image, cmap="gray"), plt.axis("off"), plt.title("Original")
plt.subplot(1, 2, 2), plt.imshow(segmented, cmap="gray"), plt.axis("off"), plt.title("Segmented")
plt.show()
```

#***5. GUI Canvas:-***

```python
import tkinter as tk

def draw(event):
    canvas.create_oval(event.x, event.y, event.x+3, event.y+3, fill="black")

def clear():
    canvas.delete("all")

root = tk.Tk()
root.title("Canvas App")

canvas = tk.Canvas(root, bg="white", width=600, height=400)
canvas.pack(fill=tk.BOTH, expand=True)
canvas.bind("<B1-Motion>", draw)

tk.Button(root, text="Clear", command=clear).pack(pady=5)

root.mainloop()
```
