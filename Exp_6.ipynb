#***Uppercase And Lowercase Alphabets In A String:-***

```python
def count_lower_upper(string):
  dlu = { 'Lower':0, 'Upper':0}
  for a in string:
    if a.islower():
      dlu['Lower']+=1
    elif a.isupper():
      dlu['Upper']+=1
  return(dlu)
A = count_lower_upper("Infinity And Beyond")
print(A)
A = count_lower_upper("I Love Python")
print(A)
```
