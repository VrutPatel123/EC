#***3. Largest Of 3 User Given Number:-***

```python
Ajay = int(input('Enter age of Ajay :'))
Ram = int(input('Enter age of Ram :'))
Shyam = int(input('Enter age of Shyam :'))
if Ajay>Ram and Ajay>Shyam:
  print('Ajay is the Oldest')
elif Ram>Ajay and Ram>Shyam:
  print('Ram is the Oldest')
else:
  print('Shyam is the Oldest')
```

#***5.Factorial Of User Given Number:-***

```python
Fact = 1
num = int(input('Positive Integer: '))
if num < 0:
    num = -num
if num == 0:
    print('Factorial of 0 is 1')
else:
    a = num
    while(a > 1):
        Fact *= a
        a -= 1
    print('Factorial of',num,'is',Fact)
```

#***6. Uppercase And Lowercase Characters In A String:-***

```python
def count(string):
    upper = 0
    lower = 0
    for i in range(len(string)):
        if string[i] >= 'A' and string[i] <= 'Z':
            upper += 1
        elif string[i] >= 'a' and string[i] <= 'z':
            lower += 1
    return {"u":upper,"l":lower}
A = input('Enter a string:')
B = count(A)
print(f'Uppercase Alphabets:{B["u"]}')
print(f'Lowercase Alphabets:{B["l"]}')
```

#***7. Sum Of Digit Of 5-Digit Number***

```python
num = int(input('Integer: '))
if num < 0: num = -num
sum = 0
while num > 0:
    sum = sum + num % 10
    num = num // 10
print('Sum of Digits =', sum)
```
