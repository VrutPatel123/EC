#***Finding The Smallest From 3 User Given Number***

```python
Hari = int(input('Enter age of Hari :'))
Ram = int(input('Enter age of Ram :'))
Krishna = int(input('Enter age of Krishna :'))
if Hari<Ram and Hari<Krishna:
  print('Hari is the Youngest')
elif Ram<Hari and Ram<Krishna:
  print('Ram is the Youngest')
else:
  print('Krishna is the Youngest')
```
