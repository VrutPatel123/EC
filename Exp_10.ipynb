#***Subplot:-***

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,4,7,10])
y = np.array([10,15,30,20])
plt.subplot(2,2,1)
plt.plot(x,y)
```

#***Scatter:-***

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,4,7,10,12,16,17])
y = np.array([10,15,30,20,25,35,40])
plt.scatter(x,y)
plt.show()
```

#***Bar:-***

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,4,7,10])
y = np.array([10,15,30,20])
plt.bar(x,y,color = 'red')
plt.show()
```
