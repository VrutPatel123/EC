#***If Else Condition:-***

```python
x = 5
y = 10
if x > y:  # NOTE:- Indentation(whitespace at the beginning of a line) is required
  print("X is greater than Y")
elif x == y:
  print("X is equal to Y")
else:
  print("Y is greater than X")
```

#***For Loop:-***

```python
n = 7
for i in range(0,n):
  print(i)
  if i >= 5:
    break
```

```python
MyCar = ["BMW","Benz","Audi","Mercedes"]
for x in MyCar:
  print(x)
```

#***While Loop:-***

```python
MyCar = ["BMW","Benz","Audi","Mercedes"]
i = 0
while i < len(MyCar):
  print(MyCar[i])
  i += 1
```
